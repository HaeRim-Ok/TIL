### 네이버 뉴스 제목 파싱

```python
import requests
from bs4 import BeautifulSoup
# 절대 경로 지정 - news.naver.com + href 속성 값
from urllib.parse import urljoin

url = 'https://news.naver.com/'
# get 요청 시 custom header 지정
req_header = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36'
}
res = requests.get(url, headers=req_header)

print(res.status_code)
#print(type(res)) # <class 'requests.models.Response'>
print(f'response header: {res.headers}')
print(f'response header: {res.request.headers}')
```

```python
#if res.ok :
if res.status_code == 200 : 
    # Response 객체에서 소스 추출 - 가독성 좋지 않음
    # 여기까지가 request 역할
    html=res.text
    #print(html)
    
    soup = BeautifulSoup(html, 'html.parser')
    #print(soup)
    
    # 특정 data 가져오기 - CSS selector
    # href의 속성 값에 'read.nhn'이 포함된 모든 데이터 추출
    atag_list = soup.select("a[href*='read.nhn']")
    #print(type(atag_list), len(atag_list))
    #print(atag_list)
    for idx, atag in enumerate(atag_list,1):
        title = atag.text.strip()
        # url 변수의 도메인 + atag의 href 속성 값
        link = urljoin(url,atag['href'])
        print(idx,title, link)
        
```

### 다음 뉴스 제목 파싱

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

url = 'https://news.daum.net/'
req_header={
    'user-agent' : ''
}
res = requests.get(url, headers = req_header)
print(res.status_code)
print(res.headers)
print(res.request.headers)
```

```python
if res.ok : 
    html = res.text
    soup = BeautifulSoup(html, 'html.parser')
    #print(soup)
    atag_list = soup.select("a[href*='news.v.daum.net']")
    #print(atag_list)
    for idx, atag in enumerate(atag_list,1):
        title = atag.text.strip()
        link = urljoin(url, atag['href'] )
        print(idx, title, link)
```
